```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV
df = pd.read_csv(r"C:\Users\ningn\Downloads\XHS KOC List - Sheet1.csv")
```

```python
df.columns
```

```python
following = df["following count"]
fans = df["fans count"]
liked = df["liked n collected"]
```

```python
for i in range(len(liked)):
    if liked[i][-1] == 'k':
        liked[i] = int(liked[i][:-1])*1000
    else: liked[i] = int(liked[i])
```

```python
plt.scatter(fans, liked)
plt.title("number ratio of like/star vs fan")
plt.xlabel("number of fans")
plt.ylabel("number of like and star")
plt.ylim(0,60000)
plt.xlim(0,2000)
plt.show()
```

```python
import seaborn as sns
def plot_distribution(data,name):
    if name == "number of liked":
        sns.histplot(data, kde=True, stat='density', bins=90)
    else:
        sns.histplot(data, kde=True, stat='density', bins=30)
    plt.title("Distribution of "+name)
    plt.xlabel(name)
    plt.ylabel('Density')
    if name == "number of liked":
        plt.xlim(0,200000)
    plt.show()
plot_distribution(following,"number of followers")
plot_distribution(fans,"number of fans")
plot_distribution(liked,"number of liked")
```
